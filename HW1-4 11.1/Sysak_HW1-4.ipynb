```python
import numpy as np
import scipy.sparse as scs
```

```python
edges = []
with open('Cit-HepTh.txt', 'r') as inf:
    for i in range(4):
        _ = inf.readline()
    for line in inf.readlines():
        edges.append([x.strip() for x in line.split('\t')])

edges = np.array(edges)
```

```python
edges.shape
```

```python
NUMtoID = []
IDtoNUM = {}
cnt = 0
for x in edges.ravel():
    if x not in IDtoNUM:
        IDtoNUM[x] = cnt
        NUMtoID.append(x)
        cnt += 1
```

```python
n = len(IDtoNUM)
```

```python
from collections import Counter
link_cnt = Counter(edges[:, 0])
```

```python
%%time
P = scs.dok_matrix((n, n))
for num, [j, i] in enumerate(edges):
    P[IDtoNUM[i], IDtoNUM[j]] = 1 / link_cnt[j] 
```

```python
P
```

```python
P = P.tocsr()
```

```python
v = np.random.rand(n)
v = v / np.linalg.norm(v, ord=2)
old_v = np.ones((n,)) * 100
i = 0

while np.linalg.norm(v - old_v, ord=2) > 0.01 and i < 1000:
    old_v = v
    plchldr = P.dot(v)
    v = plchldr / np.linalg.norm(plchldr, ord=2)
    i += 1
    print("{}: {}".format(i, np.linalg.norm(v - old_v, ord=2)))
```

```python
sorted(v / np.linalg.norm(v, ord=1), reverse=True)[:10]
```

```python
top = [NUMtoID[x] for x in np.argsort(-v)[:10]]
```

```python
top
```

```python
for i in range(len(top)):
    if len(top[i]) < 7:
        top[i] = '0' * (7 - len(top[i])) + top[i]
```

```python
years = []
for i in top:
    if int(i[:2]) > 80:
        year = '19' + i[:2]
    else:
        year = '20' + i[:2]
    years.append(year)
```

```python
for ind, year in zip(top, years):
    with open('Abstracts/{}/{}.abs'.format(year, ind), 'r') as inf:
        print(inf.read())
```
